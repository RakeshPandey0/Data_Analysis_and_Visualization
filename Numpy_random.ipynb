#### Random data distribution

Generate a 1-D array containing 100 values, where each value has to be 3, 5, 7 or 9.

The probability for the value to be 3 is set to be 0.1

The probability for the value to be 5 is set to be 0.3

The probability for the value to be 7 is set to be 0.6

The probability for the value to be 9 is set to be 0

```python
x = random.choice([3, 5, 7, 9], p=[0.1, 0.3, 0.6, 0.0], size=(100))

print(x)
```

```python
x = random.choice([3, 5, 7, 9], p=[0.1, 0.3, 0.6, 0.0], size=(3, 5))

print(x)
```

## Normal(Gaussian) Distribution

##### loc - (Mean)
##### scale - (Standard Deviation)
##### size - Shape of returned array

```python
x = random.normal(size=(2, 3))
print(x)
```

```python
x = random.normal(loc=1, scale=2, size=(2, 3))

print(x)
```

### Visualization 

```python
sns.displot(random.normal(size=1000), kind='hist')
plt.show()
```

### Binomial distribution

#### n -number of trials
#### p -probability of occurence of each trial
#### size -shape of returned array

```python
sns.displot(random.binomial(n=10, p=0.5, size=1000), kind='kde')
plt.show()
```

#### Difference between normal and binomial distribution

The main difference is that normal distribution is continuous while binomial distribution is discrete, but if there are enough data points, it will be quite similar.

```python
sns.displot(random.normal(loc=50, scale=5, size=1000), kind='kde', label='normal')
sns.displot(random.binomial(n=100, p=0.5, size=1000), kind='kde', label='binomial')
plt.show()
```

### Poisson Distribution

##### lam - rate or known number of occurrences
##### size - shape of returned array

```python
sns.displot(random.poisson(lam=2, size=1000), kind='kde')
plt.show()
```

### Uniform distribution

Here each event has equal chances of occuring

##### a- lower bound -default 0.0
##### b- upper bound -default 1.0
##### size- The shape of the returned array

```python
x= random.uniform(size=(2, 3))
print(x)
```

```python
sns.displot(random.uniform(size=1000), kind='kde')
plt.show()
```

### Logistic distribution

used to describe growth. Used extensively in machine learning in logistic regression, neural networks

##### loc- mean, where the peak is. Default 0.
##### scale- standard deviation, the flatness of distribution. Default 1.
##### size- the shape of returned array.

```python
x = random.logistic(loc=1, scale=2, size=(2, 3))
print(x)
```

```python
sns.displot(random.logistic(loc=50, scale=10, size=1000), kind='kde')
plt.show()
```

### Multinomial distribution

It is generalization of binomial distribution.

##### n - number of possible outcomes
##### pvals - list of probabilities of outcomes
##### size - shape of returned array

```python
x = random.multinomial(n=6, pvals=[1/6, 1/6, 1/6, 1/6, 1/6, 1/6,])
print(x)
```

```python
sns.displot(data=x, kind='kde')
plt.show()
```

### Exponential distribution

##### used for desctibing time till next event eg. failure/success etc.

##### scale - inverse of rate (see lam in poisson distribution) default 1.0
##### size - shape of returned array

```python
x = random.exponential(scale = 2, size=(2, 3))
print(x)
```

```python
sns.displot(random.exponential(scale=2, size=1000), kind='kde')
plt.show()
```

### Chi Square Distribution

##### df- degree of freedom
##### size- shape of returned array

```python
x = random.chisquare(df=2, size=(2, 3))
print(x)
```

```python
sns.displot(random.chisquare(df=2, size=1000))
plt.show()
```

### Rayleigh Distribution

used for signal processing

##### scale- standard deviation. Defines how flat the distribution will be
##### size- shape of the returned array

```python
x= random.rayleigh(scale=2, size=(2, 3))
print(x)
```

```python
sns.displot(random.rayleigh(scale=2, size=1000))
plt.show()
```

### Pareto Distribution

A distribution following Pareto's law i.e. 80-20 distribution (20% factors cause 80% outcome).

##### a- shape parameter
##### size- shape of returned array

```python
x= random.pareto(a=2, size=(2, 3))
print(x)
```

```python
sns.displot(random.pareto(a=2, size=1000))
plt.show()
```

### Zipf's Distribution

Zipf's law - Nth common term is the 1/n time of the most common term. E.g. the 5th most common word in English occurs nearly 1/5 times as often as the most common word.

##### a- distribution parameter
##### size- shape of returned array

```python
x= random.zipf(a=2, size=(2, 3))
print(x)
```

```python
sns.displot(random.zipf(a=2, size=100))
plt.show()
```
