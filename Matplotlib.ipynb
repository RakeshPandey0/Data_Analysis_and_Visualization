## Subplot

#### Subplot() function

#### subplot(row,  column,  figure number)

#### subplot(1, 2, 1) --> 1 row, 2 column, 1st figure
#### subplot(1, 2, 2) --> 1 row, 2 column, 2nd figure

#### subplot(2, 1, 1) --> 2 row, 1 column, 1st figure
#### subplot(2, 1, 2) --> 2 row, 1 column, 2st figure

## plotting side-by-side

```python
#Plot 1:

x = np.array([0, 1, 2, 3, 4])
y = np.array([3, 8, 1, 10, 6])
plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.title("Plot 1", loc='left')

#Plot 2

x = np.array([2, 4, 5, 7, 8])
y = np.array([12, 16, 18, 25, 30])
plt.subplot(1, 2, 2)
plt.plot(x, y)
plt.title("Plot 2", loc='left')

plt.show()
```

### Color

### Color each dot

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])
plt.scatter(x, y, c=colors)
plt.show()
```

```python
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])
plt.scatter(x, y, c=colors, cmap='viridis')
plt.show()
```

### Size

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20, 60, 100, 140, 10, 60, 45, 40, 50, 5, 55, 80, 90])

plt.scatter(x, y, s=sizes)
plt.show()
```

### Alpha --> to adjust the transparency of scatter points

```python
plt.scatter(x, y, s=sizes, alpha=0.5)
plt.show()
```

## Combining color, size and alpha

```python
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```

## Pie chart

```python
y = np.array([10, 35, 25, 30])
plt.pie(y)
plt.show()
```

#### Labels

```python
mylabels = np.array(["Apple", "Lemons", "Orange", "Banana"])
plt.pie(y, labels=mylabels)
plt.show()
```

## Start Angle

```python
plt.pie(y, labels=mylabels, startangle=90)
plt.show()
```

## Explode

```python
myexplode = np.array([0.1, 0, 0, 0.2])
plt.pie(y, labels=mylabels, explode=myexplode)
plt.show()
```

## Shadow

```python
plt.pie(y, labels=mylabels, explode=myexplode, shadow=True)
plt.show()
```

## Colors

```python
mycolors= np.array(['red', 'hotpink', 'green', 'blue'])
plt.pie(y, labels=mylabels, explode=myexplode, colors=mycolors)
plt.show()
```


'r' - Red
'g' - Green
'b' - Blue
'c' - Cyan
'm' - Magenta
'y' - Yellow
'k' - Black
'w' - White


## Legend

```python
plt.pie(y, colors=mycolors, explode=myexplode, labels=mylabels, shadow=True)
plt.legend()
plt.show()
```

```python
plt.pie(y, colors=mycolors)
plt.legend(labels=mylabels, title="For fruits")
plt.show()
```
