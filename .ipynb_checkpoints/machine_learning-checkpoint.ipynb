## Linear Regression

```python
from scipy import stats
```

```python
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]
```

```python
slope, intercept, r_value, p_value, std_error = stats.linregress(x, y)
```

```python
def regression_line(x):
    return slope*x + intercept
```

```python
mymodel = list(map(regression_line, x))
```

```python
plt.scatter(x, y)  #scatter plot of x and y values
plt.plot(x, mymodel) # plot regression line
plt.show()
```

#### correlation between x and y

```python
print(r_value)
```

#### correlation is near to -1 so there is a relatively strong relationship between x_value and y_value

## Prdicting values
#### suppose we want to predict the y value for x = 10

```python
print(regression_line(10))
```

## Polynomial regression

```python
x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]
```

```python
#creating a 3rd degree polynomial model to fit the given data
mymodel = np.poly1d(np.polyfit(x, y, 3))
```

```python
#100 equally spaced data points between 1 and 22
myline = np.linspace(1, 22, 100)
```

```python
plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.show()
```

## R-squared

##### In order to predict future values, we need to have good relationship between x and y values. For that we need r-square value. It lies between 0 and 1. 0 means no relation and 1 means 100% relationship.

##### to find r-value, we need python and sklearn

```python
from sklearn.metrics import r2_score
```

#### r2_score(y_true, y_predicted)
#### it returns relationship between true and predicted value. If relationship is good, we can use polynomial regression to predict future values.

```python
mymodel = np.poly1d(np.polyfit(x, y, 3))
```

```python
print(r2_score(y, mymodel(x)))
```

#### r2_score is 0.94 that means there is a good relationship so we can use polynomial regression to predict future values.

## Predict future values

```python
print(mymodel(17))
```

## Multiple regression

### multiple regression is like linear regression but with more than one independent variable.

```python
import pandas as pd
```

```python
df = pd.DataFrame(pd.read_csv("car_data_for_regression.csv"))
```

```python
print(df)
```

### Here weight and volume are independent variables and CO2 is dependent variable.

#### we define list of independent variable as X and list of dependent variable as y.

```python
X = df[['Weight', 'Volume']]
y = df['CO2']
```

```python
print(X)
```

```python
print(y)
```

```python
from sklearn import linear_model
X = df[['Weight', 'Volume']]
y = df['CO2']
```

First we create a linear regression object

```python
regr = linear_model.LinearRegression()
```

Now we use fit() method to provide the object with independent and dependent values. With this, the regression object is filled with data that describes the relationship between independent and dependent parameters.

With this we can use the object to predict the amount of CO2 emmission by providing value for weight and volume.

```python
regr.fit(X, y) #regr.fit(independent, dependent)
```

```python
predictedCO2 = regr.predict([[1200, 1300]])
print(predictedCO2)
```

## Coefficient

### coefficient is the factor that describes the relationship with an unknown variables.

### In this case we can see the coefficients of weight against CO2 and volume against CO2 which will tell us what would happen if we increase or decrease one of the independent values.

```python
print(regr.coef_)
```

## Result :
### weight: 0.00755095
### volume: 0.00780526
#### These values represent that when weight is increased by 1KG, the CO2 emmission will increase by 0.00755095g.
#### Similarly if volume is increased by 1m^3, the CO2 emission will increase by 0.00780526g.

#### What if we increase the weight by 1000kg.

```python
CO2_predicted = regr.predict([[2200, 1300]])
print(CO2_predicted)
```

### This shows that the coefficient is: 106.45363807 - 98.9026908 = 0.007550947

# Scale

### It is a process of bringing dataset to a similar range. It is important, specially for the algorithms that are sensitive to scale of input features.

## Primary goal:

### 1. Equalize the scale: Ensures all features have similar scale. prevents features from dominating learning process.
### 2. Stabilize the optimization process: Helps optimization algorithms to converge faster.

```python
df = pd.DataFrame(pd.read_csv("car_data_for_scaling.csv"))
```

```python
print(df)
```

#### Here, volume is given in liters instead of cm^3. (1.0 instead of 1000)

#### It is difficult to compare 1.0 to 790. So we need to scale them both into comparable values, to see how much one values is compared to another.

### To do this, we will be using standardization method. The formula is:
## z = (x-u)/s

#### z - standardized value
#### x - original value
#### u - mean
#### s - standard deviation

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
```

```python
X = df[['Weight', 'Volume']]
y = df['CO2']
scaledX = scale.fit_transform(X)
```

```python
print(scaledX)
```

### We have standardized weight and volume

# Predict CO2 value

```python
regr = linear_model.LinearRegression()
regr.fit(scaledX, y)
```

Now we standardize a random data to predict y value.

```python
scaled = scale.transform([[2300, 1.3]])
```

Now we predict CO2 value

```python
predictedCO2 = regr.predict([scaled[0]])
```

```python
print(predictedCO2)
```

#### Note: we use fit_transform when we and to both learn and apply a transformation to your data in a single step.
####     we use transform when we have already trained a transformer, such as standardization or PCA, and we want to apply transformation to a new or unseen data.

# Train/Test

### It is a method to measure the accuracy of a model. For this we split the data into two sets: training set and testing set.

### 80% for training and 20% for testing.

```python
np.random.seed(2)
x = np.random.normal(3, 1, 100) #(loc, scale, size)
y = np.random.normal(150, 40, 100) /x #(loc, scale, size)
```

```python
plt.scatter(x, y)
plt.xlabel("minutes")
plt.ylabel("money-spent")
plt.show()
```

#### The x axis represents the number of minutes before making a purchase.

#### The y axis represents the amount of money spent on the purchase.

### Split data into train and test

```python
train_x = x[:80]
train_y = y[:80]
test_x = x[80:]
test_y = y[80:]
```

### Display training set

```python
plt.scatter(train_x, train_y)
plt.show()
```

### Display testing set

```python
plt.scatter(test_x, test_y)
plt.show()
```

### Create a model using training data

```python
mymodel = np.poly1d(np.polyfit(train_x, train_y, 4))
```

```python
myline = np.linspace(0, 6, 100)
```

```python
plt.scatter(train_x, train_y)
plt.plot(myline, mymodel(myline))
plt.show()
```

```python
r2 = r2_score(train_y, mymodel(train_x))
print(r2)
```

### Now bringing in the testing data

```python
mymodel = np.poly1d(np.polyfit(test_x, test_y, 4))
```

```python
r2 = r2_score(test_y, mymodel(test_x))
print(r2)
```

## Predict values

```python
print(mymodel(5))
```
