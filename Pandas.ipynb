## Pandas Series

Pandas series is like a column in a table. It is a one-dimensional array holding data of any type

```python
a = [1, 4, 6, 8]
myvar = pd.Series(a)
print(myvar)
```

### Labels

##### if nothing is specified, the values are labeled with their index number.

```python
print(myvar[0])
```

#### with specified indices

```python
a = [1, 3, 5, 7, 9, 12, 15]
myvar = pd.Series(a, index =['a', 'b', 'c', 'd', 'e', 'f', 'g'])
print(myvar)
```

```python
print(myvar['d'])
```

#### Key/value objects as series

```python
calories = {"day1": 420, "day2":380, "day3":390}
myvar = pd.Series(calories)
print(myvar)
```

#### selectively creating a series

```python
calories = {"day1": 420, "day2":380, "day3":390}

myvar = pd.Series(calories, index=["day1", "day2"])

print(myvar)
```

## DataFrame

## Cleaning Data

Cleaning means fixing bad data. bad data could be:
##### Empty cells
##### Data in wrong format
##### Wrong data
##### Duplicates

## Cleaning Empty Cells

##### Remove cells

```python
df = pd.read_csv('date_data.csv')
print(df.to_string())
```

```python
print(df.info())
```

```python
new_df=df.dropna()
print(new_df.to_string())
```

#### if we want to make changes to original dataframe, we use inplace= True argument

```python
df=pd.read_csv('date_data.csv')
df.dropna(inplace= True)
print(df.to_string())
```

## Replace Empty values

```python
df=pd.read_csv('date_data.csv')
df.fillna(130, inplace=True)
print(df.to_string())
```

### Replace only for specified columns

```python
df=pd.read_csv('date_data.csv')
df['Calories'].fillna(130, inplace=True)
print(df.to_string())
```

### Replacing with mean, median and mode

```python
df=pd.read_csv('date_data.csv')
x=df['Calories'].mean()
df['Calories'].fillna(x, inplace=True)
print(df.to_string())
```

```python
df=pd.read_csv('date_data.csv')
x=df['Calories'].median()
df['Calories'].fillna(x, inplace=True)
print(df.to_string())
```

```python
df=pd.read_csv('date_data.csv')
x=df['Calories'].mode()[0]
df['Calories'].fillna(x, inplace=True)
print(df.to_string())
```

### Cleaning data of wrong format

```python
df = pd.read_csv('date_data.csv')
print(df.to_string())
```

```python
sd=pd.read_csv('date_data.csv')
df['Date'] = pd.to_datetime(df['Date'], format='mixed')
print(df.to_string())
```

```python
df.dropna(subset=['Date'], inplace=True)
print(df.to_string())
```

```python
df=pd.read_csv('date_data.csv')

for x in df.index:
  if df.loc[x, "Duration"] > 120:
    df.loc[x, "Duration"] = 120

print(df.to_string())
```

## Removing Rows

```python
for x in df.index:
  if df.loc[x, "Duration"] > 120:
    df.drop(x, inplace = True)

print(df.to_string())
```

## Removing Duplicates

```python
df=pd.read_csv('date_data.csv')
print(df.duplicated())
```

```python
df.drop_duplicates(inplace=True)
print(df.to_string())
```
